```python
import numpy as np
import matplotlib.pyplot as plt

# باز کردن فایل و خواندن داده‌ها
data = np.loadtxt('DUAL-optimizer.txt', delimiter=',', skiprows=0)

# استخراج ستون‌های مختلف
steps = data[:, 0]
loss = data[:, 1]
acc = data[:, 2]
learning_rate = data[:, 3]

# رسم نمودار
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Step')
ax1.set_ylabel('Loss', color=color)
ax1.plot(steps, loss, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # ایجاد یک محور جدید برای نمایش دوم
color = 'tab:blue'
ax2.set_ylabel('Accuracy', color=color)
ax2.plot(steps, acc, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # تنظیم محل نمایش نمودارها
plt.show()
```
